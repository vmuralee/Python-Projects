$\color{Blue} {S} = \text{Susceptible}$
$\color{Red} {I} = \text{infected}$
$\color{Green} {R} = \text{Recovered(Removed)}$

## SIR Model 

$\begin{array} 
&\frac{dS}{dt} &= -R_{trans}*(SI) \\ 
\frac{dI}{dt} &= R_{trans}*(SI)-R_{recov}*I\\
\frac{dR}{dt} &= R_{recov}*I\end{array}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# initial condition
Istart = 0.01
Sstart = 1-Istart
Rstart = 0
```

```python
#parameters
def Corona(x,t):
    R_trans = 1.2
    R_recov = 0.23
    maxT = 1
    S = x[0]
    I = x[1]
    R = x[2]
    dSdt = -R_trans*S*I
    dIdt = R_trans*S*I-R_recov*I
    dRdt = R_recov*I
    return [dSdt,dIdt,dRdt]
```

```python
x0 = [Sstart,Istart,Rstart]
t = np.linspace(0,22,1000)
x = odeint(Corona,x0,t)
```

```python
S = x[:,0]
I = x[:,1]
R = x[:,2]
```

```python
#No curfew
#transmision rate = 3.32 (assumed)
plt.plot(t,S,label='Suceptible')
plt.plot(t,I,label='Infected')
plt.plot(t,R,label='Recovered')
plt.xlabel('Days')
plt.ylabel('Rate')
plt.title('Without Modi put Curfew')
plt.legend()
```

```python
#After Modi put curfew
#transmision rate = 1.2 (assumed)
plt.plot(t,S,label='Suceptible')
plt.plot(t,I,label='Infected')
plt.plot(t,R,label='Recovered')
plt.xlabel('Days')
plt.ylabel('Rate')
plt.title('With Modis Curfew')
plt.legend()
```
